# torch.optim

While you can build everything yourself using these low-level objects, many people have done so before and packaged the results in reusable libraries. 

Embora seja possível construir tudo sozinho usando esses objetos de baixo nível, muitas pessoas já fizeram isso antes e empacotaram os resultados em bibliotecas reutilizáveis.

We'll first look at ```torch.optim```, which contains a number of _optimizers_. Using these, we don't have to implement the gradient descent step ourselves. This may not seem like a big part of our code, but it can get more complicated when we want to try variations on gradient descent like Adam. To illustrate, let's use the Adam optimizer in our linear regression example.

Primeiro, vamos dar uma olhada em ```torch.optim```, que contém uma série de _otimizadores_. Usando esses otimizadores, não precisamos implementar manualmente o passo de descida do gradiente. Isso pode não parecer uma parte importante do nosso código, mas pode ficar mais complicado quando queremos experimentar variações na descida do gradiente, como o algoritmo Adam. Para ilustrar, vamos usar o otimizador Adam no nosso exemplo de regressão linear.



````python
import torch
````

````python
from torch.optim import Adam 

# hiperparametros

iterations = 101
learning_rate = 2.0

# reconstruindo os dados e o modelo

x = torch.randn(1000, 32)  
wt, bt = torch.randn(32,1), torch.randn(1)

t = torch.mm(x, wt) + bt

# iniciando os pesos e bias de forma aleatórias

w, b = torch.randn(32,1, requires_grad=True), torch.randn(1, requires_grad=True)

# criando o otimizador. Para isso você precisa
# saber duas coisas 
# - o  learning_rate 
# - which parameters its responsible for

opt = Adam(lr=learning_rate, params=[w, b])

for i in range(iterations):
    
    # forward/backward, same before(igual antes)
    y = torch.mm(x, w) + b
    r = t - y
    loss = (r ** 2).mean()

    loss.backward()
    # -- Note que o otimizador não calcula os gradientes. Ainda assim fazemos isso por conta própria. O otimizador pega os gradientes e os usa para adaptar os parâmetros. 

    # print the loss
    if i % 20 == 0:
        print(f'iterations {i: 4} loss {loss:.4}')
        # {i: 4} e {loss:.4} são apenas formatação de inteiros(espaço) e float
    # Perform the gradient descent step(descidade gradiente)
    opt.step() 

    # The optimizer can zero the gradients for us 
    # (but we still have to tell it to do so)
    opt.zero_grad()
````

You may have noticed that Adam (which is supposed to be better than gradient descent in many ways), actually takes longer to converge.That's because this is a very simple problem. Adam's strength shows when you train very large networks with many weights all doing different things.

Você pode ter percebido que o Adam (que se supõe ser melhor do que a descida do gradiente de várias maneiras), na verdade, leva mais tempo para convergir. Isso ocorre porque esse é um problema muito simples. A força do Adam se mostra quando você treina redes muito grandes, com muitos pesos fazendo coisas diferentes.
